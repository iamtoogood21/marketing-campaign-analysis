```python
import pandas as pd

df = pd.read_csv("../data/marketing_data.csv")

df.head()
import matplotlib.pyplot as plt

df["response"].value_counts().plot(kind="bar")
plt.title("Campaign Response Distribution")
plt.show()
```

## Insight

Most customers did not respond to the campaign.  
This means marketing efforts need better targeting.

```python
import matplotlib.pyplot as plt

df["response"].value_counts().plot(kind="bar")
plt.title("Campaign Response Distribution")
plt.xlabel("Response (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

## Insight 1

Most customers did not respond to the marketing campaign.  
This suggests that the campaign is not effectively targeting the right audience.

```python
import matplotlib.pyplot as plt

plt.scatter(df["income"], df["response"])
plt.title("Income vs Campaign Response")
plt.xlabel("Income")
plt.ylabel("Response")
plt.show()
```

## Insight 2

Customers with higher income appear more likely to respond.  
This suggests targeting higher-income groups may improve campaign success.

```python
#  Previous Purchases
df.groupby("previous_purchases")["response"].mean().plot(kind="line")
plt.title("Previous Purchases vs Response Rate")
plt.xlabel("Previous Purchases")
plt.ylabel("Response Rate")
plt.show()
```

## Insight 3

Customers with more previous purchases are more likely to respond.  
This indicates loyal customers are a key target segment.

```python
##  Website Visits vs Response
df.groupby("website_visits")["response"].mean().plot()
plt.title("Website Visits vs Response")
plt.show()
```
